```python
import pandas as pd

df = pd.read_csv('../../imu_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
import numpy as np

df['OrientationX_sin'] = np.sin(df['OrientationX'] * np.pi / 180)
df['OrientationX_cos'] = np.cos(df['OrientationX'] * np.pi / 180)
df['OrientationY_sin'] = np.sin(df['OrientationY'] * np.pi / 180)
df['OrientationY_cos'] = np.cos(df['OrientationY'] * np.pi / 180)
df['OrientationZ_sin'] = np.sin(df['OrientationZ'] * np.pi / 180)
df['OrientationZ_cos'] = np.cos(df['OrientationZ'] * np.pi / 180)
```

```python
df.drop(columns=['OrientationX', 'OrientationY', 'OrientationZ'], inplace=True)
```

```python
df.head(10)
```

```python
dataset_dance = df[df['Set'] == 'Dance']
dataset_idle = df[df['Set'] == 'Idle']
dataset_walk = df[df['Set'] == 'Walk']
dataset_run = df[df['Set'] == 'Run']
dataset_spin = df[df['Set'] == 'Spin']
```

```python
import matplotlib.pyplot as plt

plt.bar(['Dance', 'Idle', 'Walk', 'Run', 'Spin'], [len(dataset_dance), len(dataset_idle), len(dataset_walk), len(dataset_run), len(dataset_spin)])
plt.title('Representation of the dataset (before looping)')
plt.xlabel('Label')
plt.ylabel('Number of frames')
plt.show()
```
