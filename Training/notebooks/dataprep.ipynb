```python
import pandas as pd

df = pd.read_csv('../../imu_data.csv')
```

```python
import numpy as np

df['OrientationX_sin'] = np.sin(df['OrientationX'] * np.pi / 180)
df['OrientationX_cos'] = np.cos(df['OrientationX'] * np.pi / 180)
df['OrientationY_sin'] = np.sin(df['OrientationY'] * np.pi / 180)
df['OrientationY_cos'] = np.cos(df['OrientationY'] * np.pi / 180)
df['OrientationZ_sin'] = np.sin(df['OrientationZ'] * np.pi / 180)
df['OrientationZ_cos'] = np.cos(df['OrientationZ'] * np.pi / 180)
```

```python
df.drop(columns=['OrientationX', 'OrientationY', 'OrientationZ'], inplace=True)
```

```python
df.head(10)
```

```python
dataset_dance = df[df['Set'] == 'Dance']
dataset_idle = df[df['Set'] == 'Idle']
dataset_walk = df[df['Set'] == 'Walk']
dataset_run = df[df['Set'] == 'Run']
dataset_spin = df[df['Set'] == 'Spin']
```

```python
import matplotlib.pyplot as plt

plt.bar(['Dance', 'Idle', 'Walk', 'Run', 'Spin'], [len(dataset_dance), len(dataset_idle), len(dataset_walk), len(dataset_run), len(dataset_spin)])
plt.title('Representation of the dataset (before looping/capping)')
plt.xlabel('Label')
plt.ylabel('Number of frames')
plt.show()
```

```python
# Fixed 500 frames per animation
datasets = [dataset_dance, dataset_idle, dataset_walk, dataset_run, dataset_spin]
processed_datasets = []

for dataset in datasets:
    processed_data = []
    for anim_index in dataset['AnimIndex'].unique():
        anim_data = dataset[dataset['AnimIndex'] == anim_index].iloc[1:]
        if len(anim_data) > 500:
            anim_data = anim_data.iloc[:500]
        repeat_times = 500 // len(anim_data)
        if repeat_times > 0:
            repeated_data = pd.concat([anim_data] * repeat_times, ignore_index=True)
            processed_data.append(repeated_data)
        else:
            processed_data.append(anim_data)
    if len(processed_data) > 0:
        processed_dataset = pd.concat(processed_data, ignore_index=True)
        processed_datasets.append(processed_dataset)

dataset_dance, dataset_idle, dataset_walk, dataset_run, dataset_spin = processed_datasets
```

```python
import matplotlib.pyplot as plt

plt.bar(['Dance', 'Idle', 'Walk', 'Run', 'Spin'], [len(dataset_dance), len(dataset_idle), len(dataset_walk), len(dataset_run), len(dataset_spin)])
plt.title('Representation of the dataset (after looping/capping)')
plt.xlabel('Label')
plt.ylabel('Number of frames')
plt.show()
```

```python
# add noise to the dataset
for i in range(len(processed_datasets)):
    processed_datasets[i] = processed_datasets[i].copy()
    noise_columns = [
        'OrientationX_sin', 'OrientationX_cos',
        'OrientationY_sin', 'OrientationY_cos',
        'OrientationZ_sin', 'OrientationZ_cos',
        'AccelerationX', 'AccelerationY', 'AccelerationZ',
        'GyroscopeX', 'GyroscopeY', 'GyroscopeZ'
    ]
    for col in noise_columns:
        processed_datasets[i][col] += np.random.normal(0, 0.1, len(processed_datasets[i]))
```

```python
for i in range(len(datasets)):
    processed_datasets[i].to_csv(f'../data/dataset_{i}.csv', index=False)
```
